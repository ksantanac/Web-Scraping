```python
import requests
from bs4 import BeautifulSoup
```

```python
# produto = input('Qual produto vc deja?')
produto_nome = 'Mi band 5'

url = f'https://lista.mercadolivre.com.br/{produto_nome}'

response = requests.get(url)

site = BeautifulSoup(response.text, 'html.parser')

produtos = site.find('div', attrs={'class' : 'andes-card ui-search-result ui-search-result--core andes-card--flat andes-card--padding-16'})

titulo = produtos.find('h2', attrs={'class' : 'ui-search-item__title'})

link = produtos.find('a', attrs={'class' : 'ui-search-link'})

# print(produtos.prettify())
print(f"Titulo do produto: {titulo.text}")
print(f"Link do produto: {link['href']}")
```
